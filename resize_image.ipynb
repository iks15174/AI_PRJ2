```python
from torchvision import transforms
import torchvision
import os
import torch
import matplotlib.pyplot as plt
```

```python
subject_folders = ["4class_data1", "4class_data2"]
target_folder = "4class_data_t"
classes = {
    "4class_data1" : ["glass", "metal", "paper", "plastic"],
    "4class_data2" : ["metal", "glass", "paper", "plastic"],
}
target_sub_folder =  ["glass", "metal", "paper", "plastic"] 
if not os.path.exists(target_folder):
    os.makedirs(target_folder)
    for s in target_sub_folder:
        os.makedirs(target_folder + '/' + s)
```

```python
input_size = 224
trans = transforms.Compose([
    transforms.Resize((input_size, input_size)),
    transforms.ToTensor(),
])
aug_trans1 = transforms.Compose([
    transforms.Resize((input_size, input_size)),
    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1),
    transforms.RandomHorizontalFlip(p=1),
    transforms.ToTensor(),
])
aug_trans2 = transforms.Compose([
    transforms.Resize((input_size, input_size)),
    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1),
    transforms.RandomVerticalFlip(p=1),
    transforms.ToTensor(),
])
transform = [trans, aug_trans1, aug_trans2]
```

```python
img_id = 0
for sb in subject_folders:
    cls = classes[sb]
    for trans in transform:
        train_set = torchvision.datasets.ImageFolder(sb, transform=trans)
        prev = -1
        for data in train_set:
            torch.save(data[0], '%s/%s/%s'%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
            if data[1] != prev:
                plt.imshow(data[0].permute(1, 2, 0))
                plt.show()
                print("save tensor to %s/%s/%s"%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
                prev = data[1]
            img_id += 1
    #train_set = torchvision.datasets.ImageFolder(sb,transform=trans)
    #aug_train_set1 = torchvision.datasets.ImageFolder(sb, transform=aug_trans1)
    #aug_train_set2 = torchvision.datasets.ImageFolder(sb, transform=aug_trans2)
    #prev = -1
    #for data in train_set:
    #    torch.save(data[0], '%s/%s/%s'%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
    #    if data[1] != prev:
    #        plt.imshow(data[0].permute(1, 2, 0))
    #        plt.show()
    #        print("save tensor to %s/%s/%s"%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
    #        prev = data[1]
    #    img_id += 1
#
    #prev = -1
    #for data in aug_train_set1:
    #    torch.save(data[0], '%s/%s/%s'%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
    #    if data[1] != prev:
    #        plt.imshow(data[0].permute(1, 2, 0))
    #        plt.show()
    #        print("save tensor to %s/%s/%s"%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
    #        prev = data[1]
    #    img_id += 1
#
    #prev = -1
    #for data in aug_train_set2:
    #    torch.save(data[0], '%s/%s/%s'%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
    #    if data[1] != prev:
    #        plt.imshow(data[0].permute(1, 2, 0))
    #        plt.show()
    #        print("save tensor to %s/%s/%s"%(target_folder, cls[data[1]], "img_to_tensor" + str(img_id)))
    #        prev = data[1]
    #    img_id += 1
    #    
#
```
